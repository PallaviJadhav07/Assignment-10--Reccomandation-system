```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# read dataset
books= pd.read_csv('book.csv',encoding='latin')
```

```python
# droping unwanted column
books.drop('Unnamed: 0',inplace=True,axis=1)
books
```

```python
# renaming column names
books=books.rename(columns={'User.ID':'ID', 'Book.Title':'Title', 'Book.Rating':'Rating'})
books
```

```python
# sorting id colunm from asending order
books=books.sort_values('ID')
books.head(30)
```

```python
# deleting duplicate entries
books=books.drop_duplicates()
```

```python
#duplicate entry of user with same book
books[books.duplicated(subset=["ID","Title"])]
```

```python
bookclean=books.drop_duplicates(subset=["ID","Title"])
bookclean
```

```python
#making sparse matrix
user_books = bookclean.pivot(index='ID',
                                 columns='Title',
                                 values='Rating')
```

```python
user_books
```

```python
#Impute those NaNs with 0 values
user_books.fillna(0, inplace=True)    
```

```python
user_books
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances #for cosine similiarity
from scipy.spatial.distance import cosine, correlation   # for correlation similiarity
```

```python
#to check similiarity
user_sim = 1 - pairwise_distances( user_books.values, metric='cosine')      
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df 
```

```python
#Set the index and column names to user ids bcoz above it automattically assign index no. not a user id 
user_sim_df.index = books.ID.unique()
user_sim_df.columns = books.ID.unique()
```

```python
user_sim_df.iloc[0:10, 0:10]   
```

```python
 # Replace the 1 which is self make value replace by 0...not useful for further processs
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]    
```
